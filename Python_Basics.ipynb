1. What is Python, and why is it popular?
 Python is a high-level, interpreted, general-purpose programming language. It is known for its readability and simple syntax, which resembles natural language , Large Standard Library, Versatility/General-Purpose, Extensibility etc.
2.What is an interpreter in Python
An interpreter is a program that reads and executes code line by line.i.e When you run a Python script, the interpreter converts each line of the source code into machine code (bytecode) and executes it immediately.
3. What are pre-defined keywords in Python?
Keywords are special, reserved words in Python that have pre-defined meanings and perform specific operations. They are part of the language's syntax and structure.Examples: if, else, for, while etc
4. Can keywords be used as variable names?
No, keywords cannot be used as variable names, function names, class names, or any other identifier.
5. What is mutability in Python?
Mutability refers to the ability of an object to change its state or content after it has been created.e.g., Lists, Dictionaries, Sets
6. Why are lists mutable, but tuples are immutable?
Lists are Mutable because they are designed to be dynamic, allowing for in-place modification (adding, removing, changing elements) without creating a new object. They are referenced to a memory block whose contents can change.
Tuples are Immutable because they are designed for fixed collections; their size and contents are set at creation and cannot be altered. Any attempt to change them results in an error, ensuring their contents remain constant, which makes them hashable and safer for use as dictionary keys.
7. What is the difference between “==” and “=” operators in Python?
= Checks if the objects are the exact same object in memory (i.e., they have the same memory address, or ID).
== Checks if the objects have the same value.
8. What are logical operators in Python?
Logical operators are used to combine conditional statements (Boolean expressions) to determine if a condition is True or False.
9. What is type casting in Python?
Type casting (or type conversion) is the process of converting an object of one data type into another data type.
10. What is the difference between implicit and explicit type casting?
Implicit -The Python interpreter automatically converts one data type to another, usually to prevent data loss.
Explicit-The programmer manually converts the data type using built-in functions like int(), str(), etc.
11. What is the purpose of conditional statements in Python?
Conditional statements are used to execute a block of code only if a specified condition is true. This allows a program to make decisions and execute different code paths based on the state of variables or user input.The primary conditional statements are: if, elif (else if), and else.
12. How does the elif statement work?
The elif (short for "else if") statement is used after an initial if statement to check multiple conditions sequentially.
13. What is the difference between for and while loops?
for Loop -Iterating over a sequence (like a list, tuple, or string) or a block of code a fixed number of times.The number of iterations is usually determined by the length of the sequence it's iterating over. We use it when we know how many times you need to loop.
while Loop - Executing a block of code repeatedly as long as a specified condition is true.The number of iterations is determined by the condition, which must eventually become False to avoid an infinite loop  We use it when we don't know the number of iterations beforehand.
14. Describe a scenario where a while loop is more suitable than a for loop.
A while loop is more suitable when the number of iterations is not known in advance and depends on a certain condition being met.

```python
#14. Write a program to reverse a string using a while loop.
a = "pooja"
b = ""
index = len(a) - 1

while index >= 0:
    current_char = a[index]
    b = b + current_char
    index = index - 1
```

```python
#15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
a = int(input("Enter a number:"))
factorial = 1
while a > 0:
    factorial = factorial * a
    a = a - 1
print(factorial)
```
